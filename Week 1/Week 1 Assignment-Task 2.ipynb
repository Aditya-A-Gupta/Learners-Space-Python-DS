```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv('videogamesales.csv')
```

```python
df
```

```python
df['global_sales'] = df['NA_Sales'] + df['EU_Sales'] + df['JP_Sales'] + df['Other_Sales']
df_sorted = df.sort_values(by='global_sales', ascending=False)
df_sorted
```

```python
df_grpd = df.groupby('Genre')['Name'].count()
plt.figure(figsize=(12, 8))
sns.barplot(x=df_grpd.index,y=df_grpd.values)
plt.title('Total Number of Copies Sold by Genre')
plt.xlabel('Genre')
plt.ylabel('Total Number of Copies Sold')
plt.show()
```

```python
gta = df[df['Name'].str.contains('Grand Theft Auto', case=False, na=False)]

# Select the required columns
gta_cols = gta[['Name', 'Platform', 'Year', 'EU_Sales', 'JP_Sales']]

# Calculate the sum of sales in Europe and Japan
gta_cols['EU_JP_Sales'] = gta_cols['EU_Sales'] + gta_cols['JP_Sales']

# Display the filtered DataFrame
gta_cols = gta_cols[['Name', 'Platform', 'Year', 'EU_JP_Sales']]
gta_cols
```

```python
gta_tot = df[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sum()

# Display a pie chart of the total sales
plt.figure(figsize=(4,4))
plt.pie(gta_tot,labels=gta_tot.index,autopct='%1.1f%%')
plt.title('Total Sales of Grand Theft Auto Games by Region')
plt.ylabel('')
plt.show()
```
